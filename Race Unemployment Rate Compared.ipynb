```python
Unemployment_Dictionary = {"Hispanic": "Hispanic or Latino Unemployment Rate",
                          "Black": "Black Unemployment Rate",
                          "White": "White Unemployment Rate",
                          "Asian": "Asian Unemployment Rate",
                          "All Races": "All Unemployment Rate"}
```

```python
for key, val in Unemployment_Dictionary.items():
    print(f"the key is:{key} the val is:{val}")
```

```python
race = []
month = []
year = []
rate = []
date = []
months = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]
for key, val in Unemployment_Dictionary.items():
    df = pd.read_excel(f"Resources/{val}.xlsx")
    df = df.loc[(df["Year"] == 2019) | (df["Year"] == 2020)]
#     print(df)
    for mth in range(12):
        for index, row in df.iterrows():
            month.append(months[mth])
            race.append(key)
            year.append(int(row["Year"]))
            rate.append(row[mth+1])
            date.append(datetime.datetime(int(row["Year"]),mth+1, 1))

df_global = pd.DataFrame({'Race': race,
                          'Month': month,
                          'Year': year,
                          'Rate': rate,
                          'Dates': date})
df_global = df_global.sort_values(["Race","Dates"])
df_global
```

```python
clean_df = df_global.dropna(how='any')
clean_df
```
