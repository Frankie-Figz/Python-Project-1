```python
state_dictionary = {"Alabama": "Alambama Unemployment Rate",
                   "Georgia": "Georgia Unemployement Rates"}
```

```python
for key, val in state_dictionary.items():
    print(f"the key is:{key} the val is:{val}")
```

```python
df_global = pd.DataFrame(columns = ['Series ID', 
                                   'Year',
                                   'Period',
                                   'Label',
                                   'Value',
                                   'State'])

for key, val in state_dictionary.items():
    df = pd.read_csv(f"Resources/{val}.csv")
    df['State']=key
    df_global = pd.concat([df_global, df])
```
